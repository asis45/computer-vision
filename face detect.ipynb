```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

virat=cv2.imread('C://Users//Hp//Desktop//cv//obama.jpg',0)
plt.imshow(virat,cmap='gray')

face_cascade=cv2.CascadeClassifier("C://Users//Hp//Desktop//cv//haarcascade_frontalface_default.xml")

def detect_face(img):
    
  
    face_img = img.copy()
  
    face_rects = face_cascade.detectMultiScale(face_img) 
    
    for (x,y,w,h) in face_rects: 
        cv2.rectangle(face_img, (x,y), (x+w,y+h), (255,255,255), 10) 
        
    return face_img
    
```

```python
result= detect_face(virat)
plt.imshow(result,cmap='gray')
```

```python
#real time face detection
cap=cv2.VideoCapture(0)

while True:
    ret,frame=cap.read(0)
    frame=detect_face(frame)
    
    cv2.imshow("Real time Face detection",frame)
    
    if cv2.waitKey(1)==27:
        break

cap.release()
cv2.destroyAllWindows()
```
